```python
import pandas as pd
import numpy as np
```

```python
df = pd.read_csv("/content/daily-minimum-temperatures-in-me.csv")
df.head()
```

```python
df.columns = ["month", "sales"]

df["month"] = pd.to_datetime(df["month"])
df.columns = ["month", "sales"]

# Convert date correctly
df["month"] = pd.to_datetime(df["month"])

# Clean sales
df["sales"] = pd.to_numeric(df["sales"], errors="coerce")
df["sales"] = df["sales"].fillna(method="ffill")

# Set index and frequency
df = df.set_index("month")
df = df.asfreq("MS")
```

```python
df.head()
df.index.freq
```

```python
import matplotlib.pyplot as plt

plt.figure(figsize=(8,4))
plt.plot(df["sales"])
plt.title("Monthly Shampoo Sales")
plt.xlabel("Date")
plt.ylabel("Sales")
plt.show()
import seaborn as sns

sns.histplot(df["sales"], kde=True)
plt.title("Sales Distribution")
plt.show()
```

```python
train = df.iloc[:-3]
test = df.iloc[-3:]
```

```python
from statsmodels.tsa.statespace.sarimax import SARIMAX

model = SARIMAX(
    train["sales"],
    order=(1,0,1),
    seasonal_order=(0,0,0,0)
)

result = model.fit(disp=False)
print(result.summary())
```

```python
test_forecast = result.forecast(steps=3)
test_forecast
```

```python
from sklearn.metrics import mean_absolute_error, mean_squared_error
import numpy as np

mse = mean_squared_error(test["sales"], test_forecast)
rmse = np.sqrt(mse)

mae = mean_absolute_error(test["sales"], test_forecast)
mape = np.mean(np.abs((test["sales"] - test_forecast) / test["sales"])) * 100

print(f"RMSE: {rmse:.2f}")
print(f"MAE:  {mae:.2f}")
print(f"MAPE: {mape:.2f}%")

```

```python
smape = np.mean(
    2 * np.abs(test["sales"] - test_forecast) /
    (np.abs(test["sales"]) + np.abs(test_forecast))
) * 100

print(f"sMAPE: {smape:.2f}%")
```

```python
model_202 = SARIMAX(
    train["sales"],
    order=(2,0,2),
    seasonal_order=(0,0,0,0)
)

result_202 = model_202.fit(disp=False)
pred_202 = result_202.forecast(steps=3)

rmse_202 = np.sqrt(mean_squared_error(test["sales"], pred_202))
print("RMSE (2,0,2):", rmse_202)
```

```python
from statsmodels.stats.diagnostic import acorr_ljungbox

lb_test = acorr_ljungbox(residuals, lags=[5], return_df=True)
lb_test
```
